### Question 1

```python
var('x,y')
f(x) = x^4  - 2*x^2
k(x) = abs(diff(f,x,2))/(1+(diff(f,x)^2))^(3/2)
p1 = plot(f,x,-2,2, color= "red", legend_label = "$f$")
p2 = plot(k,x,-2,2, color = "blue", legend_label= "$k$")
p1.set_legend_options(font_size = 9)
p2.set_legend_options(font_size = 9, loc = 2)

show(p1+p2, figsize = 6)
```

We see that the curvature function has its local maximums when the curve $f$ is more "bended". Note also that the sharp edges of $\kappa$ coincide with the inflection points of $f$. See why?

We only need to plot the curve whose equation in polar coordinates is $r = e^\theta$, and two of its tangents. The tangents are:

* $(x,y) = (1+t,t)$ 
* $(x,y) = (-t,e^{\pi/2} +t)$

```python
var('t')
C = polar_plot(exp(t),(t,-1,2), color = "blue")
t1 = parametric_plot([1+t,t],(t,-1,3), color = "red")
t2 = parametric_plot([-t,e^(pi/2)+t],(t,-2,3), color = "green")
show(C+t1+t2, gridlines = True, xmin = -2, xmax = 3, ymin = -1, ymax = 6)
```

```python
var('t')
b=1
a=4
x(a,b,t) = (a-b)*cos(t) + b*cos((a-b)*t/b)
y(a,b,t) = (a-b)*sin(t) - b*sin((a-b)*t/b)
```

We first fix $b=1$, and fix several values for $a = n/d$.

We notice that the shape of the curve is pretty much the same, but it gets smaller and smaller as $d$ increases. Also, for bigger values of $d$ one may want to try larger intervals for the aprameter $\theta$


**Note:** in order to use the code below, and make vary the parameters $n$ or $d$ I defined the functions $x$ and $y$ as functions of actually three variables. For this part $b=1$ and $a = 1/d$ as $d$ runs over some set of integers...

```python
b=1

D=[2,3,4,5,6,10] # The values for d
A= [1/d for d in D]
Curves = [parametric_plot([x(a,b,t),y(a,b,t)], (t,-2*pi,2*pi), plot_points= 500, 
                          legend_label = 'a=%s'%a.n(digits=2)) for a in A]

for g in Curves:
    g.set_legend_options(font_size = 8, loc = 2)
    g.fontsize(8)
    
    
Array = graphics_array(((Curves[0],Curves[1], Curves[2]),
                     (Curves[3],Curves[4], Curves[5])))

Array.show(figsize  = 10)
```

Let now $d= 2$ and $n$ run through $\{3,5,7,9,11, 12\}$. This choice guarantees that $n$ and $d$ do not have a common factor, except for the last case. We include it here to compare.

```python
b=1
d=2
N=[3,5,7,9,11,12] # The values for d
Curves = [parametric_plot([x(n/d,b,t),y(n/d,b,t)], (t,-2*pi,2*pi), plot_points= 500, 
                          legend_label = 'n=%s'%n) for n in N]
for g in Curves:
    g.set_legend_options(font_size = 8, loc = 2)
    g.fontsize(8)    
Array = graphics_array(((Curves[0],Curves[1],Curves[2]),
                     (Curves[3],Curves[4], Curves[5])))
Array.show(figsize  = 10)
```

We see that the graph looks like an star with $n$ cusps whenever $n$ and $d$ do not have a common factor. See what happens when $n=12$ and $d=2$? The number of cusps is $12/2 = 6$
